```
#T(n)=4
from time import time
def ejemplo1( n ):
    start_time = time()
    c = n + 1
    d = c * n
    e = n * n
    total = c + e - d
    elapsed_time = time() - start_time
    print("Tiempo transcurrido: %0.10f segundos." % elapsed_time)
    print(f"total={ total }")
for c in range(100,1100,100):
  print(f"Ejecucion con el numero {c} \n{ejemplo1( c )}-------------------------------------")
```

```
#T(n)=n^2+1
from time import time
def ejemplo2( n ):
    start_time = time()
    contador = 0
    for i in range( n ) :
        for j in range( n ) :
            contador += 1
    elapsed_time = time() - start_time
    print("Tiempo transcurrido: %0.05f segundos." % elapsed_time)
    return contador

for c in range(100,1100,100):
  print(f"Ejecucion con el numero {c} \n{ejemplo2( c )}-------------------------------------")
```

```
#T(n)=102
def ejemplo3 (n):
  start_time = time()
  x = n * 2
  y = 0
  for m in range (100):
    y = x-n
  elapsed_time = time() - start_time
  print("Tiempo transcurrido: %0.10f segundos." % elapsed_time)
  return y
for c in range(100,1100,100):
  print(f"Ejecucion con el numero {c} \n{ejemplo3( c )}-------------------------------------")
```

```
#T(n)=3
from time import time
def ejemplo4( n ):
    start_time = time()
    x = 3 * 3.1416 + n
    y = x + 3 * 3 - n
    z = x + y
    elapsed_time = time() - start_time
    print("Tiempo transcurrido: %0.10f segundos." % elapsed_time)
    return z

for c in range(100,1100,100):
  print(f"Ejecucion con el numero {c} \n{ejemplo4( c )}-------------------------------------")
```

```
#T(n)=1+n
from time import time
def ejemplo5( x ):
    start_time = time()
    n = 10
    for j in range( 0 , x , 1 ):
        n = j + n
    elapsed_time = time() - start_time
    print("Tiempo transcurrido: %0.10f segundos." % elapsed_time)
    return n

for c in range(100,1100,100):
  print(f"Ejecucion con el numero {c} \n{ejemplo5( c )}-------------------------------------")
```
