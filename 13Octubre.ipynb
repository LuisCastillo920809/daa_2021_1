```
#n+1+n^2
arreglo = [[1, 2, 3, 4], [ 5, 6, 7, 8]] 
total = 0
for r in range(len(arreglo)):
    for c in range(len(arreglo[r])):
        total += arreglo[r][c]
print(total+1)
```
